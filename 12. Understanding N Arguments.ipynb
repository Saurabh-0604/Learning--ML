```python
def multiply(x,y,z):
    return x*y*z
```

```python
multiply(2,3,4)
```

```python
def multiply(*numbers):
    total = 1
    for number in numbers:
        total *= number
    return total
```

```python
multiply(2,3,4,5)
```
