```python
# Lamda , Map & zip function

# lamda Anonymous func- no need to define the name

#anonymous function

# Lambda parameter : only one expression

square = lambda x : x ** 2

square(5)
```

```python
# Zip func

first_names = ["John", "Mary", "Bob"] 
last_names = ["Smith", "Jane", "Lee"]

# full_names = zip(first_names,last_names)

full_names = list(zip(first_names,last_names))

full_names
```

```python
# Map func

def square(x):
    return x ** 2

numbers = [2,3,4,5]

# squared_numbers = map(square, numbers)

squared_numbers = list(map(square, numbers))

squared_numbers
```
