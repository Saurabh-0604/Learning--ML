```python
#Classes(blueprint for creating object) -->  used to bundle data functioning together,making it easier to manage and reuse code.

# Human CLass

# John, Mary

# Attribute/ Properties - Name, age, eye color, height
# Methods/functions - Walk, Talk, Eat.
```

```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def walk(self):
        print(f"{self.name} is walking")

    def talk(self):
        print(f"{self.name} is talking")
```

```python
h1 = Human("John",27)

h1.walk()
```

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius 

    def area(self):
        """Calculate and return the area of the circle."""
        return 3.14 * (self.radius ** 2)

    def circumference(self):
        """Calculate circumferene of the circle."""
        return 3.14 * 2 * self.radius
```

```python
my_circle = Circle(5)

print(my_circle.area()) 
print(my_circle.circumference()) 
# my_circle.circumference()
# my_circle.area()
```
