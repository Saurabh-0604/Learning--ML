```python
# Arithmatic Operator

a=10
b=3

print(a+b)
print(a-b)
print(a*b)
print(a/b)  # floor division
print(a//b) # Exp
print(a**b)
```

```python
# Logical Operators

a = True
b = False

print(a and b)  # both conditions should be true
print(a or b)   # atleast one condition be true
print(not a)
```

```python
# Comparison Opeartor

a=10
b=5

print(a==b)
print(a>b)
print(a<b)
print(a>=b)
print(a<=b)
print(a!=b)
```
