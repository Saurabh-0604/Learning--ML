```python
# Conditional Statements

# if statement

temp = 25

if temp > 20:
    print("It's a warm day")
```

```python
temp = 15

if temp > 20:
    print("It's a warm day")
else:
    print("It's a cool day")
```

```python
# if- elif and else

temp = 15

if temp > 30:
    print("It's a hot day")
elif temp > 20:
    print("little warm")
else:
    print("It's a cool day")
```

```python
# Nested conditions

temp = 23
wheather = "rainy"

if temp > 15:
    if wheather == "sunny":
        print("little warm sunny day")
    else:
        print("It's warm but not sunny")
else:
    print("It's not a warm day")
```
