# Matplotlib

Matplotlib is a powerful and widely-used plotting library for Python. It provides a variety of tools for creating static, interactive, and animated visualizations in Python. Whether you are exploring data, creating reports, or developing interactive dashboards, Matplotlib offers a comprehensive set of functions to generate high-quality graphs and charts.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the data
months = np.array(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
sales = np.array([200,220,250,275,300,320,310,295,280,270,260,250])
```

```python
# 1. Basic Line Plot

plt.figure(figsize = (6,4))
plt.plot(months, sales, marker ='o',linestyle ='-', color='b')  # line plot of sales over Months
plt.title('Monthly Sales Data') # Line plot showing Sales trends over the year
plt.xlabel('Month') # Label for the x-axis
plt.ylabel('Sales') # Label for the y-axis
plt.grid(True) # Show grid lines for better readability
plt.show() # Display the plot
```

```python
# 2. Histogram

# Create a dataset for histogram
np.random.seed(0)  # For reproducibility
data = np.random.normal(loc = 0, scale = 1, size = 1000)   # Generate random data with normal distribution

plt.figure(figsize = (6,4))
plt.hist(data, bins = 30, color= 'purple', edgecolor = 'black')  # Histogram showing the distribution of data
plt.title('Random Data Distribution') # Histogram of data distribution
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# 3. Scatter Plot
# Example data for scatter plot

np.random.seed(1) # For reproducibility
x = np.random.rand(50) * 100 # Random x values
y = np.random.rand(50) * 100 # Random y values

plt.figure(figsize=(6, 4))
plt.scatter(x, y, color='red', alpha = 0.7) # Scatter plot showing relationship between x and y
plt.title('Random Scatter Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)
plt.show()
```

```python
# 4. Bar Plot
# Example data for a bar plot

categories = np.array(['A','B','C','D','E'])
values = np.array([10, 20, 15, 25, 30])

plt.figure(figsize=(6, 4))
plt.bar(categories, values, color=['blue', 'green', 'red', 'purple', 'orange']) # Bar plot showing values
plt.title('Category Values')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(False)
plt.show()
```

```python
#5. Box Plot
# Example data for box plot

np.random.seed(2) # For reproducibility
group1 = np.random.normal(loc=0, scale=1, size=100) # Data for group 1
group2 = np.random.normal(loc=1, scale=1.5, size=100) # Data for group 2
plt.figure(figsize=(6, 4))
plt.boxplot([group1, group2], tick_labels=['Group 1', 'Group 2'])  # Box plot showing distribution of data a
plt.title('Box Plot of Two Groups')
plt.xlabel('Group')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
# 6. Heatmap (Correlation Matrix)
# Create a correlation matrix for demonstration

np.random.seed(3) # For reproducibility
data_matrix = np.random.rand(10, 10) # Random data matrix
plt.figure(figsize=(6,4))
plt.imshow(data_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar (label='Value')
plt.title('Heatmap of Random Data Matrix')
plt.show()

```

```python
# Generate same simple data using arrays

x = np.linspace(0, 10, 100)  # 100 values from 0 to 10
y1 = np.sin(x) # Sine wave
y2 = np.cos(x) # Cosine wave

# Create a figure
plt.figure(figsize=(6, 2))

# First subplot: Sine wave (assigned to the first plot in the grid)
plt.subplot(2, 1, 1) # 1 row, 2 columns, first plot 
plt.plot(x, y1, 'b-') # Line plot for sine wave 
plt.title('Sine Wave') # Title of the subplot

# Second subplot: Cosine wave (assigned to the second plot in the grid)
plt.subplot(2, 1, 2) #1 row, 2 columns, second plot
plt.plot(x, y2, 'r--') # Line plot for cosine wave
plt.title('Cosine Wave') # Title of the subplot

# Adjust layout to avoid overlap
plt.tight_layout()

# Display the PLots
plt.show()
```
