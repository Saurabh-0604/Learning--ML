# Seaborn

Seaborn is a powerful and user-friendly data visualization library built on top of Matplotlib. It provides an interface for creating attractive and informative statistical graphics with fewer lines of code. Seaborn simplifies the process of creating complex visualizations and enhances Matplotlib plots with improved aesthetics and additional functionalities.

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings

warnings.filterwarnings("ignore")
```

```python
# Load the datasets 
df = pd.read_csv("C:\\Users\\Administrator\\Downloads\\sample_data (1).csv")
```

```python
df.head()
```

```python
# Set the style for plots
sns.set(style = 'whitegrid') # Use a white grid style for better visualization
```

```python
# 1. Histogram with KDE (Kernel Density Estimation)

plt.figure(figsize=(6,4))
sns.histplot(df['Spending_Score'], bins = 20, kde = True, color = 'purple')  # Histogram with KDE overlay
plt.title('Distribution of Spending Score')  # Histogram with Spending Score distribution
plt.xlabel('Spending Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Scatter Plot with Hue

plt.figure(figsize = (6,4))
sns.scatterplot(x = 'Income', y = 'Spending_Score', data = df, hue = 'Gender', palette = 'deep', alpha = 0.7)
plt.title('Income vs Spending Score')  # Scatter Plot showing Income vs Spending score with Gender
plt.xlabel('Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
# 3. Box Plot

plt.figure(figsize = (6,4))
sns.boxplot(x = 'Region', y = 'Income', data = df, palette = 'pastel' )  # Box of showing income distribution
plt.title('Income Distribution by Region')  # Box plotby Income Region
plt.xlabel('Region')
plt.ylabel('Income')
plt.show()
```

```python
# 4. Bar Plot

plt.figure(figsize=(6, 4))
sns.barplot(x='Education_Level', y='Income', data=df, estimator='mean', palette='muted')
plt.title('Average Income by Education Level') # Bar plot of Average Income by Education Level
plt.xlabel('Education Level')
plt.ylabel('Average Income')
plt.xticks(rotation=45) # Rotate x-axis Labels for better readability
plt.show()
```

```python
#5. Pair Plot

plt.figure(figsize=(5, 5))
sns.pairplot(df [['Age', 'Income', 'Spending_Score', 'Height', 'Weight', 'Gender']], diag_kind='kde', hue = 'Gender')
plt.show()
             
```

```python
# 6. Heatmap (Correlation Matrix)

plt.figure(figsize = (6,4))
corr_matrix = df[['Age','Income','Spending_Score','Height','Weight']].corr()  # Compute the correlation matrix
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm', fmt = '.2f')  # Heatmaop showing correlation between features with
plt.title('Correlation Heatmap')  # Heatmap of Correlation Matrix with annotations
plt.show()
```
