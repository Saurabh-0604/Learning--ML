What is Ndarray?

An ndarray (N- dimensional array) is a multidimensional container of items of the same type and size. It allows u to perform operations on large dataset effeciently.

```python
import numpy as np

# Creating a 1D array
array_1d = np.array([1,2,3,4,5])
print("1D Array:", array_1d)

# Creating a 2D array
array_2d = np.array([[1,2,3],[4,5,6]])
print("2D Array:\n", array_2d)
```

```python
# Creating a 2D array of all zeros

zeros_array = np.zeros((3,4))   # 3 rows and 4 column
print("Aray of Zeros:\n",zeros_array)
```

```python
# Creating a 2D array of all ones

ones_array = np.ones(5)
print("Array of ones:\n",ones_array)
```

```python
# Creating an arrays with value from 0 to 9

range_array = np.arange(10)
print("Range Array:", range_array)
```

```python
# Creating an array with 5 evenly spaced value between 0 and 1

linspace_array = np.linspace(0, 1, 5)
print("Linspace Array:",linspace_array)
```

```python
print("Shape of array_2d", array_2d.shape)
print("Size of array_2d", array_2d.size)
```
