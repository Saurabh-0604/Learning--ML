```python
import pandas as pd

# Reading the Dataset
df = pd.read_csv('C:\\Users\\Administrator\\Downloads\\inventory_data (1).csv')
df.head()
```

```python
df.info()
```

```python
print('\nMissing value in each column:')
print(df.isnull().sum())
```

```python
# Handle mising values
# Fill missing values in 'stock' with the median of column

df['quantity_in_stock'] = df['quantity_in_stock'].fillna(df['quantity_in_stock'].median())
```

```python
print('\nMissing values in each column: ')
print(df.isnull().sum())
```

```python
# Check for Duplicates 

print('\nDuplicate rows in the Dataframe: ')
print(df.duplicated().sum())
```

```python
df.info()
```

```python
# Changing the Data type of the product_id column
df['product_id'] = df['product_id'].astype(float)
```

```python
print(df['unit_price'].dtype)
```

```python
print(df['unit_price'].unique()[:10])
```

```python
# Ensure 'Stock' and 'Price' are Positive

df = df[df['quantity_in_stock'] >= 0]
df = df[df['reorder_level'] >= 0]
```
